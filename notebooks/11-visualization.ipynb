# Data Visualization with Pandas
**`11-visualization.ipynb`*

Visualization is an important step in **data analysis**.  
Pandas integrates with **Matplotlib** to quickly create charts.



---



## Step 1: Import Libraries

```python

import pandas as pd
import matplotlib.pyplot as plt

# Enable inline plotting
%matplotlib inline
```

---



## Step 2: Sample Data

```python
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [250, 300, 400, 350, 500, 600],
    "Profit": [40, 60, 80, 65, 90, 120]
}

df = pd.DataFrame(data)
print(df)

```

---



## Step 3: Line Plot


```python

df.plot(x="Month", y="Sales", kind="line", marker="o", title="Monthly Sales")
plt.show()
```


---


## Step 4: Bar Plot

```python

df.plot(x="Month", y="Profit", kind="bar", title="Monthly Profit")
plt.show()
```


---



## Step 5: Multiple Columns in One Plot

```python

df.plot(x="Month", y=["Sales", "Profit"], kind="line", marker="o", title="Sales vs Profit")
plt.show()
```


---


## Step 6: Histogram

```python

df["Sales"].plot(kind="hist", bins=5, title="Sales Distribution")
plt.show()
```


---


## Step 7: Box Plot

```python
df[["Sales", "Profit"]].plot(kind="box", title="Sales & Profit Distribution")
plt.show()
```


---


## Step 8: Scatter Plot


```python
df.plot(x="Sales", y="Profit", kind="scatter", title="Sales vs Profit")
plt.show()
```


---



## Step 9: Pie Chart


```python
df.set_index("Month")["Sales"].plot(kind="pie", autopct="%1.1f%%", title="Sales Share by Month")
plt.ylabel("")  # remove extra label
plt.show()
```


---



## Step 10: Customizing Plots


```python
ax = df.plot(x="Month", y="Sales", kind="line", marker="o", color="red", figsize=(8,5))
ax.set_xlabel("Month")
ax.set_ylabel("Sales ($)")
ax.set_title("Customized Sales Plot")
plt.show()

```


---



## Step 11: Subplots

```python
df.plot(x="Month", y=["Sales", "Profit"], kind="line", subplots=True, layout=(2,1), figsize=(8,6))
plt.show()
```


---

## Step 12: Summary

* Use `.plot(kind=...)` for quick charts
* Supported: **line, bar, histogram, box, scatter, pie**
* Customize with **Matplotlib**
* Subplots allow comparison of multiple variables

Visualization makes patterns **easier to understand** 🚀
